```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.svm import SVR
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
train_df=pd.read_csv("train.csv")
train_df.head()
```

```python
# split the data into features and target using another approach
X= train_df.iloc[:,:-1]
y=train_df["churn"]
```

```python
#encode the target
le=LabelEncoder()
y_=le.fit_transform(y)
```

```python
X= train_df.copy()
y=X.pop("churn")
```

```python
#category features encoding
X_dummies=pd.get_dummies(X,drop_first=True)
```

```python
X_dummy=pd.get_dummies(X, columns=["state", "area_code", "international_plan", "voice_mail_plan"], drop_first=True )
X_dummy
```

```python
#partition into train and test
train_x, val_x, train_y, value_y= train_test_split(X_dummies, y_, train_size=0.8, random_state=42)
```

```python
x_train, x_test, y_train, y_test= train_test_split(X_dummy, y, test_size=0.1, random_state=42)
```

```python
dt=DecisionTreeClassifier(random_state=42)
model_1= dt.fit(x_train, y_train ) 
```

```python
svm= SVC(random_state=42)
model_2=svm.fit(x_train, y_train)
```

```python
scaler = StandardScaler()
lr = make_pipeline(scaler, LogisticRegression(max_iter=200))
model_3= lr.fit(x_train, y_train)
```

```python
prediction_1= model_1.predict(x_test)
eval_1=accuracy_score(y_test, prediction_1)
eval_1_c= confusion_matrix(y_test, prediction_1)

print(f"the accuracy score by {dt} is {eval_1} \n")
print(f"the confusion matrix for {dt} is \n{eval_1_c} \n")
```

```python
prediction_2= model_2.predict(x_test)
eval_2=accuracy_score(y_test, prediction_2)
eval_2_c= confusion_matrix(y_test, prediction_2)

print(f"the accuracy score by {svm} is {eval_2} \n")
print(f"the confusion matrix for {svm} is \n{eval_2_c}")
```

```python
prediction_3= model_3.predict(x_test)
eval_3=accuracy_score(y_test, prediction_3)
eval_3_c= confusion_matrix(y_test, prediction_3)

print(f"the accuracy score by {lr} is {eval_3} \n")
print(f"the confusion matrix for {lr} is \n{eval_3_c}")
```

```python
# train with random forest
rf=RandomForestClassifier()
model_4=rf.fit(train_x, train_y)
```

```python
prediction_4=model_4.predict(val_x)
print(classification_report(value_y,prediction_4))
```

## train the model with oversampling dataset

```python
from sklearn.utils import resample
majority_class= train_df.query("churn=='no'") # separate the majority class
minority_class= train_df.query("churn=='yes'") # separate the minority class

minority_upgrade=resample(minority_class, n_samples=len(majority_class), random_state=42)
balanced_df=pd.concat([majority_class,minority_upgrade ])
```

```python
balanced_df.churn.value_counts()
```

```python
X_s= balanced_df.copy()
y_s=X_s.pop("churn")
```

```python
X_s_dummies=pd.get_dummies(X_s,drop_first=True)
```

```python
train_x, val_x, train_y, value_y= train_test_split(X_s_dummies, y_s, train_size=0.8, random_state=42)
```

```python
dt=DecisionTreeClassifier(random_state=42)
model_5= dt.fit(x_train, y_train ) 
```

```python
prediction_5= model_5.predict(x_test)
eval_5=accuracy_score(y_test, prediction_5)
eval_5_c= confusion_matrix(y_test, prediction_5)

print(f"the accuracy score by {dt} is {eval_5} \n")
print(f"the confusion matrix for {dt} is \n{eval_5_c} \n")
```

## features selection
pca (principal component analysis) is the statistical method that reduces/simplify the complexity of data set by removing some variables while retaining most important information. this is the most commonly used in practice. 

```python
from sklearn.decomposition import PCA
```

```python
merit=PCA(n_components=20, random_state=42)
new_x=merit.fit_transform(X_s_dummies)
```

```python
train_x, val_x, train_y, value_y= train_test_split(new_x, y_s, train_size=0.8, random_state=42)
```

```python
dt=DecisionTreeClassifier(random_state=42)
model_5= dt.fit(train_x, train_y ) 
```

```python
prediction_5=model_5.predict(val_x)
print(classification_report(value_y,prediction_5))
```
