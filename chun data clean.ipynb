```python
for i in train_df.select_dtypes(include=["int","float"]).columns:
    sns.boxplot(y=train_df[i])
    sns.set_palette("Set3")
    plt.title(f"Distribution of {i}", fontsize=24, color="red")
    plt.xticks(rotation=30, color ="red")
    plt.yticks(rotation=30, color ="red")
    plt.show()
```

```python
sns.boxplot(x=train_df["churn"], y=train_df["total_day_charge"])
plt.grid(axis="y")
sns.set_palette("Set3")
plt.title(f"Distribution of total daily charge by churn", fontsize=24, color="red")
plt.xticks(rotation=30, color ="red")
plt.yticks(rotation=30, color ="red")
plt.show()
```

```python
sns.boxplot(x=train_df["churn"], y=train_df["total_day_calls"])
plt.grid(axis="y")
sns.set_palette("Set3")
plt.title(f"Distribution of total daily calls by churn", fontsize=24, color="red")
plt.xticks(rotation=30, color ="red")
plt.yticks(rotation=30, color ="red")
plt.show()
```

```python
state_count=train_df['state'].value_counts().head(10)
```

```python
sns.set_palette("Greys")
fig, ax=plt.subplots(figsize=(10,6))
state_count.plot(kind='bar')
plt.grid(axis="y")
plt.title(f"Top 10 States", fontsize=24, color="red")
plt.xticks(rotation=30, color ="red")
plt.yticks(rotation=30, color ="red")
plt.show()
```

```python
def cat_cat(col):
    sns.set_palette("Greys")
    fig, ax=plt.subplots(figsize=(10,6))
    cat_cat=pd.crosstab(
        index=train_df["churn"],
        columns= train_df[col],
        values=train_df["churn"],
        aggfunc="count")
    cat_cat.plot(kind="bar", ax=ax)
    plt.grid(axis="y")
    plt.title(f"count of {col} by churn", fontsize=24, color="red")
    plt.xticks(rotation=30, color ="red")
    plt.yticks(rotation=30, color ="red")
    plt.show()
```

```python
cat_cat("total_intl_calls")
```

```python
cat_cat("state")
```
